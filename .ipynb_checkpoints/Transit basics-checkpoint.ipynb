```python
import numpy as np
import matplotlib.pyplot as pl
import astropy.units as u
from astropy import constants as ascon

import os
import math

os.umask(0o002)
os.nice(19)
```

```python
def calc_dep(Rp, Rs): #Rp_earth, Rsun
    Rp = Rp * ascon.R_earth
    Rs = Rs * ascon.R_sun
    return (Rp/Rs)**2
```

```python
calc_dep(1, 1) *100
```

```python
calc_dep(10, 1) *100
```

```python
calc_dep(11, 1)
```

```python
def calc_T14(P_orb, a_Rs, k, b, i_deg):
    i = np.deg2rad(i_deg)
    numerator = np.sqrt((1 + k)**2 - b**2)
    denominator = a_Rs * np.sin(i)
    T14_days = (P_orb/np.pi) * np.arcsin(numerator/denominator)
    return T14_days * 24  # hours

a_Rs_earth = ascon.au/ascon.R_sun
```

```python
a_Rs_earth = (ascon.au/ascon.R_sun).value
calc_T14(365, a_Rs_earth, 0.008, 0, 90)
```

```python
calc_T14(1, 10, 0.01, 0, 90)
```
