```python
from astropy.io import fits
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
tessfilename = "https://archive.stsci.edu/missions/tess/tid/s0044/0000/0004/3422/6736/tess2021284114741-s0044-0000000434226736-0215-s_lc.fits"
```

```python
tessfitsFile = fits.open(tessfilename)
tessheader = tessfitsFile[0].header
tessdata = tessfitsFile[1].data
```

```python
tessBJD=tessdata.field("time") 
SAPflux=tessdata.field("SAP_FLUX")
SAPfluxerr=tessdata.field("SAP_FLUX_ERR")
PDCSAPflux=tessdata.field("PDCSAP_FLUX")
PDCSAPfluxerr=tessdata.field("PDCSAP_FLUX_ERR")
```

```python
npPDCSAPflux=[]
nptessBJD = []
npPDCSAPfluxerr = []

for j in range(len(tessBJD)):
    if (not np.isnan(PDCSAPflux[j]) and not np.isnan(tessBJD[j]) and not np.isnan(PDCSAPfluxerr[j])):
        npPDCSAPflux.append(PDCSAPflux[j])
        nptessBJD.append(tessBJD[j])
        npPDCSAPfluxerr.append(PDCSAPfluxerr[j])
PDCSAPflux_mean=np.mean(np.array(npPDCSAPflux))
```

```python
time = np.array(nptessBJD)
flux = np.array(npPDCSAPflux)/PDCSAPflux_mean
ferr = np.array(npPDCSAPfluxerr)/PDCSAPflux_mean
```

```python
T0_tr = 2457062.57935-2457000
P_tr = 3.484552
dr_tr = 0.79/24.
```

```python
n_tran_start=np.ceil((time[0]-T0_tr)/P_tr)
n_tran_end=np.floor((time[-1]-T0_tr)/P_tr)
```

```python
np.arange(n_tran_start, n_tran_end, 1)
```

```python
fig, ax = plt.subplots(figsize=(20,9))
plt.title('TESS Sector 44 data of K2-25 (' + str(tessheader['OBJECT']) + ')', fontsize=24)
plt.xlabel('BJD (-2457000)',fontsize=24)
plt.ylabel('Normalized flux',fontsize=24)
plt.tick_params(labelsize=24)
#plt.errorbar(time,flux,yerr=ferr,fmt=".k",capsize=0, alpha=0.5)
plt.plot(time,flux,"o", color='k', alpha=0.8)

for n_tran in np.arange(n_tran_start, n_tran_end+1, 1):
    plt.axvspan(T0_tr+P_tr*n_tran-dr_tr,T0_tr+P_tr*n_tran+dr_tr, color='tomato', alpha=0.2)

#plt.xlim(2502, 2505)
plt.ylim(0.95, 1.2)
fig.savefig('LC_K2-25.png')
```
