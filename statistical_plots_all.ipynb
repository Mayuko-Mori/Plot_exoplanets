### luminosity情報のある星が少ない！ => Stefan-Boltzmannから出す
本当はSEDでFbolを出すと一番いいらしい

```python
from astropy import constants as ascon
ascon.sigma_sb
```

```python
sb = ascon.sigma_sb.value
Rsun = ascon.R_sun.value
Lsun = ascon.L_sun.value

df_all['st_lum_calc']=np.log10(4 * np.pi  * (df_all['st_rad']*Rsun)**2 * sb * df_all['st_teff']**4  /Lsun)
```

```python
pl.plot(df_all['st_teff'], df_all['st_lum_calc'], 'o')
pl.plot(df_all['st_teff'], df_all['st_lum'], '.')
```

# multi-planetary systems

```python
# トランジット惑星が6個以上のものを選定したい
df_tran_multi= df_transit[df_transit['sy_pnum']>=6].sort_values(by='sy_pnum', ascending=True).reset_index(drop=True)
name_multi = df_tran_multi['hostname'].drop_duplicates().tolist()
len(name_multi)
```

```python
# b (通常) と f (Kepler-80 用) をまとめて作る
df_inner = df_tran_multi[
    ((df_tran_multi['hostname'] != 'Kepler-80') & (df_tran_multi['pl_letter'] == 'b')) |
    ((df_tran_multi['hostname'] == 'Kepler-80') & (df_tran_multi['pl_letter'] == 'f'))
]

b_porb = (
    df_inner
    .groupby('hostname')['pl_orbper']
    .min()
    .rename('b_orbper')
)


# hostごとの行（惑星数は同一hostで同じ想定なので first）
host_key = (
    df_tran_multi.groupby('hostname', as_index=False)['sy_pnum']
    .first()
    .merge(b_porb.reset_index(), on='hostname', how='left')
)

# 惑星数でソート → 同数なら b の周期が短い順（bが無い host は最後へ）
host_key = host_key.sort_values(
    by=['sy_pnum', 'b_orbper'],
    ascending=[True, True],
    na_position='last'
).reset_index(drop=True)

name_multi = host_key['hostname'].tolist()
```

```python
name_multi
```

```python
df_sel = df_all[df_all['hostname'].isin(name_multi)].copy()

df_sel['hostname'] = pd.Categorical(
    df_sel['hostname'],
    categories=name_multi,
    ordered=True
)

df_sel = df_sel.sort_values('hostname').reset_index(drop=True)
```
